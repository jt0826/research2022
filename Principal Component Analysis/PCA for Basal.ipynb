```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
#transpose matrix to have sample as row, gene in column
Basal_transpose = pd.read_csv(r"C:\Users\javte\Documents\SSEF project\Normalized expression data set\basal_NT_logCPMrmBatch_20221128.csv", index_col=[0])
Basal = Basal_transpose.transpose()
Basal.head()
```

```python
#PCA
pca = PCA(n_components=50)
principalComponents_Basal = pca.fit_transform(Basal)
principal_Basal_Df = pd.DataFrame(data = principalComponents_Basal
             , columns = ["PC"+str(i) for i in range(1, 51)])
```

```python
# Take pam50 column from metadata and set as index of PCA dataframe
unsorted_metadata = pd.read_csv(r"C:\Users\javte\Documents\SSEF project\Normalized expression data set\metadata\selected_metadat_20221128.csv")
unsorted_metadata.head()
Basal_metadata = unsorted_metadata.query('pam50 == "Basal" or pam50 == "Healthy"')
Basal_metadata.head()
Basal_pam50 = Basal_metadata[('pam50')].to_list()
principal_Basal_Df['pam50'] = Basal_pam50
principal_Basal_Df.set_index('pam50', drop = True, inplace = True)
principal_Basal_Df.head()
```

```python
#formatting plot
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('PC1',fontsize=20)
plt.ylabel('PC2',fontsize=20)
plt.title("Basal PCA",fontsize=20)
targets = ['Healthy', 'Basal']
colours = ['r', 'g']
#colour code points according to pam50
for target, colour in zip(targets,colours):
    index = principal_Basal_Df.index == target
    plt.scatter(principal_Basal_Df.loc[index, 'PC1']
               , principal_Basal_Df.loc[index, 'PC2'], c = colour, s = 50)
#plot legend
plt.legend(targets,prop={'size': 15})



```
