```python
import pandas as pd
```

```python
er=pd.read_csv(r"C:\Users\javte\Documents\SSEF project\datasets\scRNA\epithelial\de_markers_ERepi_wilcox.csv",index_col=[0])
her2=pd.read_csv(r"C:\Users\javte\Documents\SSEF project\datasets\scRNA\epithelial\de_markers_HER2epi_wilcox.csv",index_col=[0])
tn=pd.read_csv(r"C:\Users\javte\Documents\SSEF project\datasets\scRNA\epithelial\de_markers_TNepi_wilcox.csv",index_col=[0])
```

```python
epithelial_dataset = pd.read_csv(r"C:\Users\javte\Documents\SSEF project\datasets\scRNA\epithelial\raw_cts_allEpi.csv",index_col=[0])
epithelial_dataset=epithelial_dataset.transpose()
display(epithelial_dataset.head()); epithelial_dataset.shape
```

```python
intersection_genes=list(set(er.index)&set(her2.index)&set(tn.index))
display(str(intersection_genes)); display(len(intersection_genes))
epithelial_dataset_filtered = epithelial_dataset[intersection_genes]
```

```python
from sklearn import preprocessing
standardized_epithelial = preprocessing.scale(epithelial_dataset)
standardized_epithelial = pd.DataFrame(standardized_epithelial,index=list(epithelial_dataset.index)
                                       ,columns=list(epithelial_dataset.columns))
display(standardized_epithelial.head()); standardized_epithelial.shape
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA()
principalComponents = pca.fit_transform(standardized_epithelial)
principalDf = pd.DataFrame(data=principalComponents
                           ,columns=["PC"+str(i) for i in range(1,len(principalComponents)+1)])
```

```python
epithelial_target = pd.read_csv(r"C:\Users\javte\Documents\SSEF project\datasets\scRNA\epithelial\Epi_sampInfo.csv",index_col=[0])
display(epithelial_target.head()); epithelial_target.shape
```

```python
#plot first 2 PCs
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=list(principalDf['PC1']),y=list(principalDf['PC2']),hue=epithelial_target['Condition'],legend='full',palette='Spectral')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
n=10
plt.bar(x=principalDf.columns[:n],height=100*pca.explained_variance_ratio_[:n])
cumulative = np.cumsum(100*pca.explained_variance_ratio_[:n])
plt.plot(principalDf.columns[:n], cumulative)
plt.rcParams["figure.figsize"] = (10,5)
```

```python
X = standardized_epithelial
#define dictionary for mapping
conditionMap = dict({'N':'0','ER':'1','HER2':'2','TN':'3'})
y = epithelial_target.Condition.replace(conditionMap)
```

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
rfc = RandomForestClassifier(random_state=77, n_estimators='auto', max_depth=5, n_jobs=-1)
boruta_selector = BorutaPy(estimator=rfc, n_estimators='auto', verbose=2, random_state=77, max_iter=100, alpha=0.0001)
boruta_selector.fit(np.array(X), np.array(y))
```

```python
boruta_genes = pd.DataFrame({'Feature':list(X.columns),'Ranking':boruta_selector.ranking_}).query('Ranking == 1')
boruta_genes.to_csv(r'C:\Users\javte\Documents\SSEF project\results\scRNA\epithelial\boruta_genes.csv')
```
